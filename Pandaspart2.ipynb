```python
def currency(x):
  if "$" in x:
    return "USD"
  else:
    return "Unknown"

df['Currency']=df['Price'].apply(currency)
#df['Currency']=df['Price'].apply(lambda x:'USD' if '$' in x else 'Unknown')
df
```

```python
df['Price']=df['Price'].str.replace('[$,.]','', regex=True) #[\D]
df
```

#Missing data

```python
df=pd.read_csv('/car-sales-missing-data.csv')
df
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#deleting nulls
df.dropna(subset=['Price'],inplace=True)
```

```python
df #As you can see, rows 7 and 8 were deleted because of NaN values.
```

```python
df.isnull().sum()
```

```python
df['Odometer'].fillna(df['Odometer'].mean(),inplace=True) #2,5,6 are the same values,we filled them with average of Odometer
df
```

```python
df['Colour']=df['Colour'].fillna(df['Colour'].mode()[0]) # we filled the colour raw with the colour that
# code sees firs and it is White ([0])
df
```

```python
for col in ['Make','Colour']:
  df[col]=df[col].fillna(df[col].mode()[0])
df
```
